```r
df <- read.csv("measurements_1553.csv", header=T)
```

```r
library (ggplot2)
library(dplyr)
```

```r
ggplot(df, aes(x=Size, y=Time, color=Type)) + geom_point() + geom_line()
```

```r
df %>% group_by(Size, Type) %>% summarise(MeanTime=mean(Time)) %>% 
    ggplot(aes(x=Size, y=MeanTime, color=Type)) + geom_point() + geom_line() + theme_bw()
```
